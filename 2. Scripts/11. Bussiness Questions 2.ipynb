```python
#4. Create a loyalty flag for existing customers using the transform() and loc() functions
# Create a max_order column
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Change head printing max
pd.options.display.max_rows = None
```

```python
# Check the head
ords_prods_merge.head(100)
```

```python
# Create condition 1
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
# Create condition 2
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
# Create condition 3
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Check frequency
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
# Return the first 60 rows that only include 3 columns
ords_prods_merge[['user_id', 'loyalty_flag', 'order_number']].head(60)
```

```python
#5. Use the loyalty flag you created and check the basic statistics of the product prices for each loyalty category (Loyal Customer, Regular Customer, and New Customer)

ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

#### This shows that there is a difference in the prices of products purchased by loyal customers vs regular or new customers. Loyal customers spend, on average, the least amount of money on products. 

```python
#6. Create a spending flag for each user based on the average price across all their orders
# Group
ords_prods_merge.groupby('department_id').agg({'order_number': ['mean']})
```

```python
# Transform
ords_prods_merge['average_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Check output
ords_prods_merge.head(20)
```

```python
# Create condition 1
ords_prods_merge.loc[ords_prods_merge['average_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
# Create condition 2
ords_prods_merge.loc[ords_prods_merge['average_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
# Check frequency
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
# Return the first 25 rows
ords_prods_merge[['user_id', 'loyalty_flag', 'order_number']].head(25)
```

```python
#7. Create an order frequency flag that marks the regularity of a users ordering behavior according to the median in the days_since_prior_order column
# Transform
ords_prods_merge['median_order_frequency']= ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Check output
ords_prods_merge.head(20)
```

```python
# Create condition 1
ords_prods_merge.loc[ords_prods_merge['median_order_frequency'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
```

```python
# Create condition 2
ords_prods_merge.loc[(ords_prods_merge['median_order_frequency'] > 10) & (ords_prods_merge['median_order_frequency'] <= 20), 'order_frequency_flag'] = 'Regular customer'
```

```python
# Create condition 2
ords_prods_merge.loc[ords_prods_merge['median_order_frequency'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
# Check frequency
ords_prods_merge['order_frequency_flag'].value_counts(dropna = False)
```

```python
# Return the first 25 rows
ords_prods_merge[['user_id', 'order_frequency_flag', 'order_number']].head(25)
```
